**Reading Images**

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

cv2 uses BGR color format and matplotlib uses RGB color format. To convert BGR to RGB, we use cv2.cvtColor function.

```python
img = cv2.imread("cat1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
```

Image Properties 

```python
print("Type:", type(img)) 
print("Shape Of Image:", img.shape) 
print("Total Numbe Of Pixels:", img.size) 
print("Image Data Type:", img.dtype) 
print("Dimension:", img.ndim)
```

Resizing an Image

```python
width = int(img.shape[1] * 70/ 100)
height = int(img.shape[0] * 70/100)
dim = (width, height)

re_img = cv2.resize(img, dim)
print('Original Image Dimensions : ',img.shape)
print('Downscaled Image Dimensions : ',re_img.shape)
plt.imshow(re_img)
plt.axis('off')
```

```python
width = int(img.shape[1] * 240/ 100)
height = int(img.shape[0] * 240/100)
dim = (width, height)

re_img = cv2.resize(img, dim)
print('Original Image Dimensions : ',img.shape)
print('Upscaled Image Dimensions : ',re_img.shape)
plt.imshow(re_img)
plt.axis('off')
```

Splitting the colours into R, G and B

```python
r,g,b = cv2.split(img)
print(img.shape)
print(r.shape)
print(r)
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.axis('off')
plt.imshow(r,cmap="Reds")
plt.subplot(132)
plt.axis('off')
plt.imshow(g,cmap="Greens")
plt.subplot(133)
plt.imshow(b,cmap="Blues")
plt.axis('off')
```

Converting into Grey scale image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img,'gray')
plt.axis('off')
```

Converting into HSV image

```python
img_hsv=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
plt.axis('off')
```

Converting into Binary image

```python
ret,thresh = cv2.threshold(gray_img,120,255,0)
plt.imshow(thresh,cmap = 'gray')
plt.title('Binary Image')
plt.axis('off')
```
